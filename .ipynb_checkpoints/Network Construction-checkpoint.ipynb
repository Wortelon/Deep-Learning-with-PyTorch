# Building a Network by Inheriting the Module Class

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class NeuralNetwork(nn.Module):
    def __init__(self, input_size=784, hidden_size=256, num_classes=10):
        # Call the superclass constructor
        super(NeuralNetwork, self).__init__()
        
        # Define network layers
        self.fc1 = nn.Linear(input_size, hidden_size)  # First fully connected layer
        self.bn1 = nn.BatchNorm1d(hidden_size)        # Batch normalization layer 1
        self.fc2 = nn.Linear(hidden_size, hidden_size) # Second fully connected layer
        self.bn2 = nn.BatchNorm1d(hidden_size)        # Batch normalization layer 2
        self.fc3 = nn.Linear(hidden_size, num_classes) # Output layer
        
        # Define dropout layer to prevent overfitting
        self.dropout = nn.Dropout(0.5)
    
    def forward(self, x):
        # Forward propagation process
        x = x.view(-1, 784)  # Flatten the input tensor
        
        # First hidden layer: linear transformation -> batch normalization -> ReLU activation -> dropout
        x = self.fc1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.dropout(x)
        
        # Second hidden layer: linear transformation -> batch normalization -> ReLU activation -> dropout
        x = self.fc2(x)
        x = self.bn2(x)
        x = F.relu(x)
        x = self.dropout(x)
        
        # Output layer: linear transformation
        x = self.fc3(x)
        
        return x

# Create model instance
model = NeuralNetwork()

# Print model architecture
print(model)    
```

# Building a Network Using the Sequential Container

```python
import torch
import torch.nn as nn

# Define the neural network using Sequential container
model = nn.Sequential(
    # Flatten the input tensor from [batch_size, 1, 28, 28] to [batch_size, 784]
    nn.Flatten(),
    
    # First hidden layer with linear transformation followed by ReLU activation
    nn.Linear(784, 256),
    nn.BatchNorm1d(256),
    nn.ReLU(),
    nn.Dropout(0.5),
    
    # Second hidden layer with linear transformation followed by ReLU activation
    nn.Linear(256, 256),
    nn.BatchNorm1d(256),
    nn.ReLU(),
    nn.Dropout(0.5),
    
    # Output layer with linear transformation (no activation for raw logits)
    nn.Linear(256, 10)
)

# Print the model architecture
print(model)

# Example usage:
# input_tensor = torch.randn(1, 1, 28, 28)
# output = model(input_tensor)   
```

# Build a Classic CNN

```python
import torch.nn as nn
import torch.nn.functional as F
 # Define the ClassicCNN Model
class ClassicCNN(nn.Module):
    def __init__(self):
        super(ClassicCNN, self).__init__()
 # Define the Network Layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)        # First convolutional layer: input channels = 3 (RGB), output channels = 32
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)       # Second convolutional layer: input channels = 32, output channels = 64
        self.pool = nn.MaxPool2d(2, 2)                                 # Max pooling layer with kernel size 2x2 and stride of 2
        self.fc1 = nn.Linear(64 * 56 * 56, 512)                        # First fully connected layer: from 64 * 56 * 56 to 512 neurons
        self.dropout = nn.Dropout(0.5)                                 # Dropout layer to reduce overfitting: drops 50% of the neurons randomly
        self.fc2 = nn.Linear(512, 1)                                   # Final output layer: maps 512 features to 1 output neuron (binary classification)

# Define the Forward  Function
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 64 * 56 * 56)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.sigmoid(self.fc2(x))
        return x
```

# Build a Simplified VGG

```python
class SimplifiedVGG(nn.Module):
    def __init__(self):
        super(SimplifiedVGG, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(128, 256, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(256, 512, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(512, 512, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(512 * 7 * 7, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)
```

# Build a Simplified AlexNet

```python
class SimplifiedAlexNet(nn.Module):
    def __init__(self):
        super(SimplifiedAlexNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=11, stride=4, padding=2), nn.ReLU(), nn.MaxPool2d(3, 2),
            nn.Conv2d(64, 192, kernel_size=5, padding=2), nn.ReLU(), nn.MaxPool2d(3, 2),
            nn.Conv2d(192, 384, kernel_size=3, padding=1), nn.ReLU(),
            nn.Conv2d(384, 256, kernel_size=3, padding=1), nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(3, 2),
        )
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((6, 6)),
            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(256 * 6 * 6, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)
```
