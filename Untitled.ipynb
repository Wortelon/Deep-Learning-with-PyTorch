1. Prerequisites: Install PyTorch & Dependencies​  
PyTorch is easier to debug and aligns with Python’s "intuitive" design. Install it via Terminal (works for Windows/macOS/Linux):  
Step 1: Install Core Libraries

# Install PyTorch (auto-detects OS/GPU; for CPU-only, use "cpuonly" instead of "cu118")​
pip3 install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118​
​
# Install supporting libraries (image processing, web framework, etc.)​
pip install pillow flask pandas matplotlib numpy
